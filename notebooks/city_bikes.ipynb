Put your parsed results into a DataFrame.

```python
city_bike_network1 = pd.DataFrame()

for station in response['network']['stations']:
    tmp_df = pd.DataFrame()
    # This df will be constructed into the df that will be stacked onto 'berlin_compiled'
    tmp_df = pd.concat([
        tmp_df,
        pd.Series(station['latitude'],name='lat'),
        pd.Series(station['longitude'],name='long'),
        pd.Series(station['extra']['slots'],name='total_bikes')
    ],
    axis=1 # horizontal concat
    )
    city_bike_network1 = pd.concat([city_bike_network1,tmp_df]) 
city_bike_network1.describe()
city_bike_network1.total_bikes.sum()
```
