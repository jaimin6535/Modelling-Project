Join the data from Part 1 with the data from Part 2 to create a new dataframe.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
result_1 = pd.read_csv("Location_data2h.csv")
print(result_1)
result_2 = pd.read_csv("Part2.csv")
print(result_2)
```

```python
df_all_rows = pd.concat([result_1, result_2])
df_all_rows

```

```python
df_all_rows=df_all_rows.reset_index(drop=True)

# or, alternatively, there's the `ignore_index` option in the `pd.concat()` function:
df_all_rows = pd.concat([result_1, result_2], ignore_index=False, axis = 1)

df_all_rows
```

```python
df_all_rows.dropna() #remove the null value
df_all_rows.head()
```

Provide a visualization that you used as part of your EDA process. Explain the initial pattern or relationship you discoved through this visualization. 

```python
fig,ax = plt.subplots()
ax.hist(df_all_rows['total_bikes'])

ax.set_title('Histogram of bikes')
ax.set_xlabel('bikes')
ax.set_ylabel('count')
plt.show()
#This graph shows hightest number of bike is nearly 300
```

```python
sns.set_theme()
sns.histplot(data=df_all_rows,x='total_bikes',y='distance')
plt.show()
##This graph shows majority bikes are near by the restaurant
```

```python
sns.set_theme()
sns.histplot(data=df_all_rows,x='total_bikes',y='restaurant_name')
plt.show()
# #This graph shows number of bikes between 10 to 30 are around all the restaurant
```

```python
sns.boxplot(data=df_all_rows,y='total_bikes',whis=2.5)
plt.show()
```

```python
sns.boxplot(data=df_all_rows,x='total_bikes',whis=2.5)
plt.show()
```

```python
sns.kdeplot(data=df_all_rows,x='total_bikes',fill =True)
plt.show()
```

```python
sns.jointplot(data=df_all_rows,y='total_bikes',x='distance')
plt.show()
```
